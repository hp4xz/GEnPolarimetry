```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
df = pd.read_csv('GEn-RUN-LIST-4-pass-kinematics.csv')

```

# Format

#### [0] = RunNumber
#### [1] = Date
#### [2] = Start Time
#### [3] = End Time
#### [4] = Target
#### [5] = Beam Current
#### [20] = Charge on Target

```python
current=np.empty(0)
dates=np.empty(0,dtype=str)
startDates=np.empty(0)
endDates=np.empty(0)
runs=np.empty(0)
start=np.empty(0)
charge=np.empty(0)

for i in range(0,len(df)):
    try:
        current=np.append(current,df.iloc[i][5])
        dates=np.append(dates,df.iloc[i][1]+" "+df.iloc[i][2]+" "+df.iloc[i][3])
        
        test=dates[i].split(" ")[0]+" " +dates[i].split(" ")[1]
        test=datetime.strptime(test,"%m/%d/%y %H:%M")
        startDates=np.append(startDates,test)
        
        test=dates[i].split(" ")[0]+" " +dates[i].split(" ")[2]
        test=datetime.strptime(test,"%m/%d/%y %H:%M")
        endDates=np.append(endDates,test)
        
        runs=np.append(runs,df.iloc[i][0])

        charge=np.append(charge,df.iloc[i][20])
    except: 
        print("failed", i)
        

removeI=np.empty(0)
for i,j in enumerate(charge):
    try:
        x=float(j)
    except:
        removeI=np.append(removeI,i)
    if j=='nan':
        removeI=np.append(removeI,i)

charge=np.delete(charge,removeI.astype(int))
```
