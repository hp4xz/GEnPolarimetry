```python
from functions import *
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from math import *
from datetime import datetime
from scipy.signal import find_peaks
```

# My Fits

```python
#My fits 
myFits=np.transpose(np.load('NMRDataFits/HunterKin2.npy'))

#Organize chronologically
dates=myFits[:,0].astype(int).astype(str)
date_objects = [datetime.strptime(date, "%Y%m%d%H%M%S") for date in dates]
dates_array = np.array(date_objects)
sorted_dates = np.sort(dates_array)
index_array = [date_objects.index(date) for date in sorted_dates]
myFitsInOrder = np.transpose(np.take(myFits, index_array, axis=0))
```

## Hunter Kin2

### -Pumping Chamber to Down Stream Target Chamber

```python
pc=myFitsInOrder[1]
ds=myFitsInOrder[3]
```

```python
ratio1=pc/ds
indices=np.empty(0)
junk = np.logical_or(np.isnan(ratio1), np.isinf(ratio1))
filteredRatio1=ratio1[~junk]
```

```python
x,y,z=plt.hist(filteredRatio1,100,range=(0.1,.4))
peaks, properties = find_peaks(x)
```
