```python
def GrabSettingsAtIndex(j):
    import os
    import numpy as np
    import pandas as pd
    import re
    NMRFolders=os.listdir('NMRData')
    
    name='NMRData/'+NMRFolders[j]+'/Downstream Coil/Settings.dat'
    with open(name, 'r') as f:
        # read the contents of the file into a list
        lines = f.readlines()

    # remove newline characters from each element
    lines = [line.strip() for line in lines]

    # print the list

    result=[]
    for string in lines[4:11]:
        # split the string into key and value
        key, value = string.split(':')
        # strip leading and trailing whitespace from key and value
        key = key.strip()
        value = value.strip()
        # convert value to integer
        value = float(value)
        # add key and value to result list
        result.append([key, value])
    print(lines[15])
    return result



 

    

```

```python
def GetSettings(indicesInRange,datesInRange):
    
        holdingField=np.empty(0)
        theta=np.empty(0)
        phi=np.empty(0)
        highField=np.empty(0)
        sweepSpeed=np.empty(0)
        sitTime=np.empty(0)
        rfFrequency=np.empty(0)
        allSettings=np.empty(0)
        for i,j in enumerate(indicesInRange):
            try:
                time=datesInRange[i].replace('_','')
                #print(int(time))
                holdingField=GrabSettingsAtIndex(j)[0][1]
                theta=GrabSettingsAtIndex(j)[1][1]
                phi=GrabSettingsAtIndex(j)[2][1]
                highField=GrabSettingsAtIndex(j)[3][1]
                sweepSpeed=GrabSettingsAtIndex(j)[4][1]
                sitTime=GrabSettingsAtIndex(j)[5][1]
                rfFrequency=GrabSettingsAtIndex(j)[6][1]
                allSettings=np.append(allSettings,[int(time),holdingField,theta,phi,highField,sweepSpeed,sitTime,rfFrequency],axis=0)
                #print(holdingField, theta, phi, highField,sweepSpeed,sitTime,rfFrequency)
                #print(allSettings[i])
            except:
                print("No Settings File")
        return allSettings
```

# Error Propagation

### PC

```python
pcust = (pcusx**2+pcusy**2)**(.5)
pcdst = (pcdsx**2+pcdsy**2)**(.5)
pc=(pcust+pcdst)/2

dsust = (dsusx**2+dsusy**2)**(.5)
dsdst = (dsdsx**2+dsdsy**2)**(.5)
ds=(dsust+dsdst)/2
    
usust = (ususx**2+ususy**2)**(.5)
usdst = (usdsx**2+usdsy**2)**(.5)
us=(usust+usdst)/2
```

```python
plt.hist(ds[np.where(ds<20)])
```

```python
len(pcusx)
```

```python
e1pcusx=pcusx**2*sqrt(2)*abs(pcusxE/pcusx)
e1pcusy=pcusy**2*sqrt(2)*abs(pcusyE/pcusy)
e1pcus_xandy=(e1pcusx**2+e1pcusy**2)**(.5)
e1pcust=pcust*sqrt(.5)*e1pcus_xandy/(pcusx**2+pcusy**2)


e1pcdsx=pcdsx**2*sqrt(2)*abs(pcdsxE/pcdsx)
e1pcdsy=pcdsy**2*sqrt(2)*abs(pcdsyE/pcdsy)
e1pcds_xandy=(e1pcdsx**2+e1pcdsy**2)**(.5)
e1pcdst=pcdst*sqrt(.5)*e1pcds_xandy/(pcdsx**2+pcdsy**2)

pcE=(e1pcust**2+e1pcdst**2)**(.5)


pcE_refined=pcE[np.where(pcE<3)]
pc_refined=pc[np.where(pcE<3)]
np.mean(pcE_refined/pc_refined)
```

### DS

```python
e1dsusx=dsusx**2*sqrt(2)*abs(dsusxE/dsusx)
e1dsusy=dsusy**2*sqrt(2)*abs(dsusyE/dsusy)
e1dsus_xandy=(e1dsusx**2+e1dsusy**2)**(.5)
e1dsust=dsust*sqrt(.5)*e1dsus_xandy/(dsusx**2+dsusy**2)


e1dsdsx=dsdsx**2*sqrt(2)*abs(dsdsxE/dsdsx)
e1dsdsy=dsdsy**2*sqrt(2)*abs(dsdsyE/dsdsy)
e1dsds_xandy=(e1dsdsx**2+e1dsdsy**2)**(.5)
e1dsdst=dsdst*sqrt(.5)*e1dsds_xandy/(dsdsx**2+dsdsy**2)

dsE=(e1dsust**2+e1dsdst**2)**(.5)


dsE_refined=dsE[np.where(dsE<3)]
ds_refined=ds[np.where(dsE<3)]
np.mean(dsE_refined/ds_refined)
```

```python
#write function to get average temperature for 15 seconds above and below given point
def GrabAround(date,data):
    from datetime import datetime, timedelta
    import os
    import numpy as np
    refined=np.empty(0)
    inRange=np.empty(0)
    inRangeIndex=np.empty(0)

    useDate=datetime.strptime(date, "%Y%m%d_%H%M%S")
    start = useDate - timedelta(minutes=2)
    end = useDate + timedelta(minutes=2)
    yep=np.empty(0)
    h=0
    ydat=data[1].astype(float)
    #print(ydat[0])
    for i in range(0,len(data[1])):
        thisDate = datetime.strptime(data[0][i], "%Y%m%d_%H%M%S")
        
        if start <= thisDate <= end:
            #print('found ',i)
            yep=np.append(yep,ydat[i])
    #print(yep)
            
        
    
    return(np.mean(yep))
```

```python
def FitLorentzian(n,q,oneSweep,i):
    error=0
    import numpy as np
    from scipy.optimize import curve_fit
    
    # Define the Lorentzian function
    def lorentzian(x, ampl, center, fwhm):
        return ampl / (1 + ((x - center) / (0.5 * fwhm))**2)
    x_data = oneSweep[n][0]
    y_data = oneSweep[n][q]
    
    
    ymean=np.mean(y_data)
    if(abs(ymean)>20):
        y_data-=ymean
        
        
    ymax=np.amax(y_data)
    ymaxarg=np.argmax(y_data)
    ymin=np.amin(y_data)
    yminarg=np.argmin(y_data)
    
    if abs(ymax)>abs(ymin):

        initial_params=[ymax,x_data[ymaxarg],1]
    else:
        initial_params=[ymin,x_data[yminarg],1]



    # Fit the Lorentzian function to the data
    try:
        
        params, cov = curve_fit(lorentzian, x_data, y_data,p0=initial_params)

        # Extract the height from the fit parameters
        height = params[0]
        
    except:
        params=[0,0,0]
        height=0
       
        error=1
    # Print the height
    #print("Height =", height)
    return(height,params)
```
