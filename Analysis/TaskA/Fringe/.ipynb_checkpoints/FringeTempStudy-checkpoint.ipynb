```python
def GrabAround(date,data):
    from datetime import datetime, timedelta
    import os
    import numpy as np
    refined=np.empty(0)
    inRange=np.empty(0)
    inRangeIndex=np.empty(0)

    useDate=datetime.strptime(date, "%Y%m%d_%H%M%S")
    start = useDate - timedelta(minutes=5)
    end = useDate + timedelta(minutes=5)
    yep=np.empty(0)
    h=0
    ydat=data[1].astype(float)
    #print(ydat[0])
    for i in range(0,len(data[1])):
        if len(data[0][i])>15:
            use=data[0][i][:15]
        else:
            use=data[0][i]
        use=use.replace("T","_") 
        thisDate = datetime.strptime(use, "%Y%m%d_%H%M%S")
        
        if start <= thisDate <= end:
            #print('found ',i)
            yep=np.append(yep,ydat[i])
    #print(yep)
            
        
    
    return(yep) 
```

```python
def GrabAroundNMR(date,data):
    from datetime import datetime, timedelta
    import os
    import numpy as np
    refined=np.empty(0)
    inRange=np.empty(0)
    inRangeIndex=np.empty(0)

    useDate=datetime.strptime(date, "%Y%m%d_%H%M%S")
    start = useDate - timedelta(minutes=15)
    end = useDate + timedelta(minutes=15)
    yep=np.empty(0)
    h=0
    ydat=data[1].astype(float)
    #print(ydat[0])
    for i in range(0,len(data[1])):
        if len(data[0][i])>15:
            use=data[0][i][:15]
        else:
            use=data[0][i]
        use=use.replace("T","_") 
        thisDate = datetime.strptime(use, "%Y%m%d_%H%M%S")
        
        if start <= thisDate <= end:
            #print('found ',i)
            yep=np.append(yep,ydat[i])
    #print(yep)
            
        
    
    return(np.mean(yep))
```
